```python
import requests
from bs4 import BeautifulSoup
```

```python
# 뉴스 1개 크롤링
url = "https://www.huffingtonpost.kr/news/articleView.html?idxno=216015"
html_doc = requests.get(url).text
html_doc
```

```python
soup = BeautifulSoup(html_doc, 'html.parser')
```

```python
soup
```

```python
title1 = soup.find('h', class_ = 'headline-title')
title1
```

```python
title1.text
```

```python
body1 = soup.find('div', class_ = 'dtale-body font-size')
```
